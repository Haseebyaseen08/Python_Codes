```python
import pandas as pd
df=pd.read_csv("irisDataset.csv")
df.head()                        #Load File
```

```python
Input=df.drop("Output",axis="columns")
Target=df.drop("Val1",axis="columns")            #Separate input output
Target=Target.drop("Val2",axis="columns")
Target=Target.drop("Val3",axis="columns")
Target=Target.drop("Val4",axis="columns")
```

```python
Input
```

```python
Target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_gender=LabelEncoder()
```

```python
Target["Output"]=le_gender.fit_transform(Target["Output"])    #Label encode 
Target.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
final_train,final_test,target_train,target_test=train_test_split(Input,Target,test_size=0.2) #Train test split
```

```python
final_train
```

```python
final_test
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#Adding input layer and 1st hidden layer
classifier=Sequential()
#Create input layer
classifier.add(Dense(4,input_shape=(4,),activation="relu"))
#Add 1st hidden layer
classifier.add(Dense(27,activation='relu'))          #Create model
#Add 2nd hidden layer
classifier.add(Dense(9,activation='relu'))
#Add output layer
classifier.add(Dense(1,activation='sigmoid'))
```

```python
classifier.compile(optimizer=keras.optimizers.SGD(learning_rate=0.1),loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
classifier.summary()
```

```python
classifier.fit(final_train,target_train,batch_size=50,epochs=150)
```

```python
outputs=classifier.predict(final_test)
```
