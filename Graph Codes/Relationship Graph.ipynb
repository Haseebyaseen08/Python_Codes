```python
class Graph:
    def __init__(self):
        self.graph={
            "Alice":[("friend of","Bob")],
            "Bob":[("friend of","Alice"),("friend of","James"),("friend of","Zach")],
            "James":[("friend of","Bob"),("friend of","Zach"),("friend of","Grace"),("friend of","Charlie"),("boss of","Zach"),("colleague of","Charlie"),("colleague of","Davina")],
            "Charlie":[("friend of","James"),("friend of","Davina"),("colleague of","James"),("colleague of","Davina")],
            "Zach":[("friend of","Alice"),("friend of","Bob"),("friend of","James"),("boss of","James"),("loves","Harriet"),("loves","Ingrid")],
            "Grace":[("friend of","James"),("married to","Fred")],
            "Davina":[("friend of","Charlie"),("friend of","Edward"),("colleague of","Charlie"),("colleague of","James")],
            "Harriet":[("loves","Zach"),("dislikes","Ingrid")],
            "Ingrid":[("loves","Zach"),("dislikes","Harriet")],
            "Fred":[("married to","Grace"),("friend of","Edward")],
            "Edward":[("friend of","Fred"),("friend of","Davina")]
        }
```

```python
def show_graph(Graph):
    for val in Graph.graph:
        for relations in Graph.graph[val]:
            print(f"\n{val} {relations[0]} {relations[1]}")
```

```python
show_graph(Graph())
```
